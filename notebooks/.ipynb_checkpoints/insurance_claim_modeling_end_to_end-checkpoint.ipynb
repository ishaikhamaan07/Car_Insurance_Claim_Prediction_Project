```python
# ==== 4. Guess target and basic schema ====
def guess_target(frame):
    candidates = [c for c in frame.columns if c.lower() in {"target","claim","made_claim","is_claim","has_claim","claim_flag","outcome"}]
    if candidates:
        return candidates[0]
    binary_like = [c for c in frame.columns if frame[c].dropna().nunique() == 2]
    if binary_like:
        for c in binary_like:
            vals = set(frame[c].dropna().unique().tolist())
            if vals <= {0,1} or vals <= {False,True}:
                return c
        return binary_like[0]
    return None

if TARGET is None:
    TARGET = guess_target(df)
    print("Guessed TARGET:", TARGET)

assert TARGET is not None, "Could not infer TARGET. Please set TARGET in the config cell."

date_like = [c for c in df.columns if any(tok in c for tok in ["date","_at","_dt"])]
for c in date_like:
    try:
        df[c] = pd.to_datetime(df[c], errors="coerce")
    except Exception:
        pass

numeric_cols = df.select_dtypes(include=["number"]).columns.tolist()
cat_cols = df.select_dtypes(include=["object","category","bool"]).columns.tolist()
dt_cols = df.select_dtypes(include=["datetime64[ns]","datetime64[ns, UTC]"]).columns.tolist()

for lst in (numeric_cols, cat_cols, dt_cols):
    if TARGET in lst:
        lst.remove(TARGET)

print("Target:", TARGET)
print("Numeric cols:", numeric_cols[:10], "..." if len(numeric_cols)>10 else "")
print("Categorical cols:", cat_cols[:10], "..." if len(cat_cols)>10 else "")
print("Datetime cols:", dt_cols)
```

```python
# ==== 10. Threshold tuning by expected cost ====
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.calibration import calibration_curve
import matplotlib.pyplot as plt

# Define misclassification costs
COST_FP = 1.0
COST_FN = 5.0

# Predicted probabilities on validation set
val_proba = best_model.predict_proba(X_val)[:, 1]

# Function to compute expected cost
def expected_cost(y_true, y_score, thr, c_fp=COST_FP, c_fn=COST_FN):
    y_pred = (y_score >= thr).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    return fp * c_fp + fn * c_fn

# Threshold grid
ths = np.linspace(0.01, 0.99, 99)
costs = [expected_cost(y_val, val_proba, t) for t in ths]

# Best threshold
best_idx = np.argmin(costs)
best_thr = ths[best_idx]
print(f"Best threshold by expected cost: {best_thr:.3f} (min cost={costs[best_idx]:.2f})")

# Calibration curve
prob_true, prob_pred = calibration_curve(y_val, val_proba, n_bins=10, strategy="quantile")

plt.figure(figsize=(6,5))
plt.plot(prob_pred, prob_true, marker="o", label="Validation")
plt.plot([0, 1], [0, 1], linestyle="--", color="gray", label="Perfectly calibrated")
plt.title("Calibration Curve (Validation)")
plt.xlabel("Mean predicted probability")
plt.ylabel("Fraction of positives")
plt.legend()
plt.show()

# Expected cost vs threshold
plt.figure(figsize=(6,5))
plt.plot(ths, costs, label="Expected cost")
plt.axvline(best_thr, color="red", linestyle="--", label=f"Best threshold ({best_thr:.3f})")
plt.title("Expected Cost vs Threshold")
plt.xlabel("Threshold")
plt.ylabel("Expected cost")
plt.legend()
plt.show()
```

```python
# ==== 12. Permutation Importance (on validation set) ====
from sklearn.inspection import permutation_importance
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Compute permutation importance
perm = permutation_importance(
    best_model, X_val, y_val,
    n_repeats=10, random_state=42, scoring="average_precision"
)

# Extract feature names from preprocessing pipeline
def get_feature_names(ct):
    names = []
    for name, trans, cols in ct.transformers_:
        if name == "remainder":
            continue
        if hasattr(trans, "named_steps"):
            # look into last step
            for step in reversed(trans.named_steps.values()):
                if hasattr(step, "get_feature_names_out"):
                    base = step.get_feature_names_out(cols)
                    break
            else:
                base = cols
        elif hasattr(trans, "get_feature_names_out"):
            base = trans.get_feature_names_out(cols)
        else:
            base = cols
        # ensure list-like
        if isinstance(base, (list, np.ndarray)):
            names.extend([f"{name}__{c}" for c in base])
        else:
            names.append(f"{name}__{base}")
    return np.array(names)

# Match feature names with permutation importances
feat_names = get_feature_names(best_model.named_steps["pre"])
k = min(len(feat_names), len(perm.importances_mean))

imp = pd.DataFrame({
    "feature": feat_names[:k],
    "imp_mean": perm.importances_mean[:k],
    "imp_std": perm.importances_std[:k],
}).sort_values("imp_mean", ascending=False)

display(imp.head(25))

# Plot top 20 features with error bars
topn = 20
plt.figure(figsize=(8,6))
plt.barh(
    imp.head(topn)["feature"][::-1],
    imp.head(topn)["imp_mean"][::-1],
    xerr=imp.head(topn)["imp_std"][::-1],
    alpha=0.7
)
plt.xlabel("Mean Importance ± STD")
plt.title(f"Permutation Importance (top {topn})")
plt.tight_layout()
plt.show()
```

```python
# ==== 14. Slice-based error analysis ====
from sklearn.metrics import average_precision_score
import pandas as pd

def slice_metric(frame, by_col, proba, y_true, min_count=10):
    out = []
    # ensure alignment
    frame = frame.loc[y_true.index]
    s_proba = pd.Series(proba, index=y_true.index)
    
    for level, idx in frame.groupby(by_col).groups.items():
        s_true = y_true.loc[idx]
        vals = s_proba.loc[idx]
        if s_true.nunique() > 1 and len(s_true) > min_count:
            out.append({
                by_col: level,
                "count": len(s_true),
                "pr_auc": average_precision_score(s_true, vals)
            })
    
    return pd.DataFrame(out).sort_values("pr_auc", ascending=False)

# Choose up to 5 categorical/bool columns
slice_cols = [c for c in X_val.select_dtypes(include=["object","category","bool"]).columns.tolist()][:5]

val_proba = best_model.predict_proba(X_val)[:, 1]

for c in slice_cols:
    print(f"\nSlice analysis by: {c}")
    display(slice_metric(X_val, c, val_proba, y_val))
```

```python
# ==== 16. Inference utility ====
import joblib
import pandas as pd
import numpy as np

def add_domain_features(frame: pd.DataFrame) -> pd.DataFrame:
    f = frame.copy()
    if {"claim_amount","premium"}.issubset(f.columns):
        f["claim_to_premium"] = (f["claim_amount"] / f["premium"].replace({0: np.nan})).fillna(0)
    if {"age","years_with_company"}.issubset(f.columns):
        f["age_per_tenure"] = (f["age"] / f["years_with_company"].replace({0: np.nan})).fillna(0)
    return f

def load_model(path: str):
    try:
        bundle = joblib.load(path)
    except Exception as e:
        raise RuntimeError(f"Failed to load model from {path}: {e}")
    return bundle["pipeline"], bundle["threshold"], bundle["target"]

def predict_from_csv(model_path: str, csv_path: str, save_path: str = None) -> pd.DataFrame:
    pipe, thr, tgt = load_model(model_path)
    
    try:
        df_new = pd.read_csv(csv_path)
    except Exception as e:
        raise RuntimeError(f"Failed to read CSV {csv_path}: {e}")
    
    # Separate target if present
    if tgt in df_new.columns:
        X_new = df_new.drop(columns=[tgt], errors="ignore")
    else:
        X_new = df_new
    
    # Add engineered features
    X_new = add_domain_features(X_new)
    
    # Align columns with training features if available
    if hasattr(pipe, "feature_names_in_"):
        X_new = X_new.reindex(columns=pipe.feature_names_in_, fill_value=0)
    
    # Predict
    proba = pipe.predict_proba(X_new)[:, 1]
    pred = (proba >= thr).astype(int)
    
    # Append predictions
    out = df_new.copy()
    out["pred_proba"] = proba
    out["pred_label"] = pred
    
    # Optionally save
    if save_path:
        out.to_csv(save_path, index=False)
        print(f"✅ Predictions saved to {save_path}")
    
    return out

# Example usage:
# preds = predict_from_csv("artifacts/best_pipeline_rf.joblib", "new_policy_data.csv", "preds.csv")
# preds.head()
```

```python
# ==== 17. Basic leakage checks (heuristic) ====
def leakage_checks(df, target, corr_thr=0.98, high_card_thr=0.9):
    suspect = []

    for c in df.columns:
        if c == target:
            continue
        
        try:
            # Numeric leakage check via correlation
            if pd.api.types.is_numeric_dtype(df[c]):
                corr = abs(df[[c, target]].corr(numeric_only=True).loc[c, target])
                if corr > corr_thr:
                    suspect.append({"feature": c, "reason": f"High correlation ({corr:.3f})"})
            
            # High-cardinality categorical check
            elif pd.api.types.is_object_dtype(df[c]) or pd.api.types.is_categorical_dtype(df[c]):
                uniq_ratio = df[c].nunique() / len(df)
                if uniq_ratio > high_card_thr:
                    suspect.append({"feature": c, "reason": f"High cardinality ({uniq_ratio:.2%})"})
        
        except Exception as e:
            print(f"⚠️ Skipped {c} due to error: {e}")

    # If nothing suspicious found → return empty DF with proper columns
    if not suspect:
        return pd.DataFrame(columns=["feature", "reason"])
    return pd.DataFrame(suspect).sort_values("reason", ascending=False)

# Run leakage check
suspect_df = leakage_checks(df, TARGET)
if suspect_df.empty:
    print("✅ No strong leakage features detected.")
else:
    print("⚠️ Potential leakage features found:")
    display(suspect_df)
```
